### Model Selection <a id='model_selection'></a>

Our business problem is to successfully predict if a customer will churn. We need to consider what our model should focus on in terms of a performance metric.

We understand that when the model predicts that a customer will churn, they should then be targeted with the appropriate strategy to reduce their chance of churning.

We want to consider the *cost* of when the model predicts a False Positive and False Negative.

**False Positive**

The model would classify this customer as likely to churn, the business would then target the customer with the strategy to prevent churn. Though if this client wasn't going to churn, the targeted strategy will most likely have little impact. 

**False Negative**

The model wouldn't classify this customer as likely to churn, the business would do nothing, the customer may proceed to churn. The business has now lost a customer, without any intervention.

The cost of applying an approved strategy to a customer would be much smaller than the cost of losing the customer. Based on this, we should be looking to build a model that minimises the False Negatives. The performance metric for this is **Recall**. We will be using this metric to first determine which is the best performing model without any tuning or adjustment to the data.

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
```

```python
## RANDOM OVER SAMPLER

# Create Base Learners
base_learners = [
                 ('XGBoost', xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')),
                 ('AdaBoost', AdaBoostClassifier())
                ]

# Create list of all models to be tested

models = [
    ('Logistic Regression', LogisticRegression()),
    ('Decision Tree', DecisionTreeClassifier(random_state=0)),
    ('Random Forest', RandomForestClassifier(random_state=0)),
    ('Support Vector Machine', SVC()),
    ('K-Nearest Neighbors', KNeighborsClassifier()),
    ('GaussianNB', GaussianNB()),
    ('AdaBoost', AdaBoostClassifier()),
    ('XGBoost', xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')),
    ('CatBoost', CatBoostClassifier(random_seed=0, verbose=False)),
    ('ExtraTree', ExtraTreesClassifier()),
    ('Stacking', StackingClassifier(estimators=base_learners, 
                                    final_estimator=CatBoostClassifier(random_seed=0, verbose=False)))
    
]

# Dictionary to store model and recall score

models_scores = {}

for model in models:
    
    # Steps for pipeline
    steps = [('scaler', StandardScaler()), ('ros', ros), model]
    
    # Create Pipeline
    pipeline = Pipeline(steps)
    
    # Train model
    model_trained = pipeline.fit(X_train, y_train)
    
    # Predictions
    y_pred = model_trained.predict(X_test)
    
    # Store model and test recall in dictionary
    models_scores[model[0]] = [np.round(metrics.recall_score(y_test, y_pred), 2),
                               np.round(metrics.accuracy_score(y_test, y_pred), 2),
                               np.round(metrics.precision_score(y_test, y_pred), 2),
                               np.round(metrics.roc_auc_score(y_test, y_pred), 2),
                               np.round(metrics.f1_score(y_test, y_pred), 2),
                               ]
```

```python
# Convert dictionary to dataframe

model_scores_df = pd.DataFrame.from_dict(models_scores, orient='index', 
                                         columns=['Recall', 'Accuracy', 'Precision', 'AUC', 'F1'])

# Sort Values

model_scores_df = model_scores_df.sort_values(by='Recall', ascending=False)

display(model_scores_df)
```

```python
## BASELINE

# Dictionary to store model and recall score

models_scores = {}

for model in models:
    
    # Steps for pipeline
    steps = [('scaler', StandardScaler()), model]
    
    # Create Pipeline
    pipeline = Pipeline(steps)
    
    # Train model
    model_trained = pipeline.fit(X_train, y_train)
    
    # Predictions
    y_pred = model_trained.predict(X_test)
    
    # Store model and test recall in dictionary
    models_scores[model[0]] = [np.round(metrics.recall_score(y_test, y_pred), 2),
                               np.round(metrics.accuracy_score(y_test, y_pred), 2),
                               np.round(metrics.precision_score(y_test, y_pred), 2),
                               np.round(metrics.roc_auc_score(y_test, y_pred), 2),
                               np.round(metrics.f1_score(y_test, y_pred), 2),
                               ]
    
# Convert dictionary to dataframe

model_scores_df = pd.DataFrame.from_dict(models_scores, orient='index', 
                                         columns=['Recall', 'Accuracy', 'Precision', 'AUC', 'F1'])

# Sort Values

model_scores_df = model_scores_df.sort_values(by='Recall', ascending=False)

display(model_scores_df)
```

```python
## RANDOM UNDERSAMPLER

from imblearn.under_sampling import RandomUnderSampler 
rus = RandomUnderSampler(random_state=0)

# Dictionary to store model and recall score

models_scores = {}

for model in models:
    
    # Steps for pipeline
    steps = [('scaler', StandardScaler()), ('rus', rus), model]
    
    # Create Pipeline
    pipeline = Pipeline(steps)
    
    # Train model
    model_trained = pipeline.fit(X_train, y_train)
    
    # Predictions
    y_pred = model_trained.predict(X_test)
    
    # Store model and test recall in dictionary
    models_scores[model[0]] = [np.round(metrics.recall_score(y_test, y_pred), 2),
                               np.round(metrics.accuracy_score(y_test, y_pred), 2),
                               np.round(metrics.precision_score(y_test, y_pred), 2),
                               np.round(metrics.roc_auc_score(y_test, y_pred), 2),
                               np.round(metrics.f1_score(y_test, y_pred), 2),
                               ]
    
# Convert dictionary to dataframe

model_scores_df = pd.DataFrame.from_dict(models_scores, orient='index', 
                                         columns=['Recall', 'Accuracy', 'Precision', 'AUC', 'F1'])

# Sort Values

model_scores_df = model_scores_df.sort_values(by='Recall', ascending=False)

display(model_scores_df)
```

```python
## SMOTE

from imblearn.over_sampling import SMOTEN
sampler = SMOTEN(random_state=0)

# Dictionary to store model and recall score

models_scores = {}

for model in models:
    
    # Steps for pipeline
    steps = [('scaler', StandardScaler()), ('sampler', sampler), model]
    
    # Create Pipeline
    pipeline = Pipeline(steps)
    
    # Train model
    model_trained = pipeline.fit(X_train, y_train)
    
    # Predictions
    y_pred = model_trained.predict(X_test)
    
    # Store model and test recall in dictionary
    models_scores[model[0]] = [np.round(metrics.recall_score(y_test, y_pred), 2),
                               np.round(metrics.accuracy_score(y_test, y_pred), 2),
                               np.round(metrics.precision_score(y_test, y_pred), 2),
                               np.round(metrics.roc_auc_score(y_test, y_pred), 2),
                               np.round(metrics.f1_score(y_test, y_pred), 2),
                               ]
    
# Convert dictionary to dataframe

model_scores_df = pd.DataFrame.from_dict(models_scores, orient='index', 
                                         columns=['Recall', 'Accuracy', 'Precision', 'AUC', 'F1'])

# Sort Values

model_scores_df = model_scores_df.sort_values(by='Recall', ascending=False)

display(model_scores_df)
```

### Cross Validation

Applying cross validation to the top 4 models above (CatBoost, Stacking, XGBoost, AdaBoost)

```python
# Create Dictionary to store models

models = {
        'CatBoost': CatBoostClassifier(random_seed=0, verbose=False),
        'XGBoost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
        'ExtraTree': ExtraTreesClassifier(),
        'Support Vector Machine': SVC(),
        'Stacking': StackingClassifier(estimators=base_learners, 
                                    final_estimator=CatBoostClassifier(random_seed=0, verbose=False))
}

# Loop through each model, printing average recall score for 5 fold cross validation

for model in models:
    
    # Create Pipeline
    
    pipeline = Pipeline([('transformer', StandardScaler()), ('ros', ros), ('estimator', models[model])])
    
    # Cross Validation - 5 Folds
    
    print(f'{model} - 5 Fold CV - Average F1')
    print(np.mean(cross_val_score(pipeline, X_train, y_train, cv=5, scoring='f1')))
    print('*'*50)
```

Based on the above scores, we will make use of CatBoost as the model for this business problem. The use of cross validation gives us more confidence that CatBoost works across all segments of the data, and the initial score on the first train set wasn't just a one-off random score.

### Model Building

Based on the above work, we'll be using CatBoost as our model to predict whether or not a customer will churn. This next section will be building it out entirely, including further fine tuning to improve the recall score.

CatBoost has some useful features that we can make use of, including an Overfitting detector, which will stop training the model once overfitting has been detected, and then pick the best model.

```python
# Set Parameters

params = {
        'random_seed': 0,
        'verbose': 200,
        'iterations': 1500,
         }

# Model

cb = CatBoostClassifier(**params)
#cb = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

# Pipeline

pipeline = Pipeline([('transformer', StandardScaler()), ('ros', ros), ('estimator', cb)])

# Fit

pipeline.fit(X_train, y_train)
```
